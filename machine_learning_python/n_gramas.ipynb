# N-gramas del libro "Don Quijote de la Mancha"

## Módulos

```python
import nltk
from nltk.util import ngrams #bigrams, trigrams
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
import requests
```

## Descargar libro "Don Quijote de la Mancha"

```python
url = "https://www.gutenberg.org/files/2000/2000-0.txt"
libro = requests.get(url).text.lower()
libro
```

## Tokenizar: convertir texto a lista de palabras

```python
nltk.download("punkt_tab")
tokens = word_tokenize(libro, language="spanish")
tokens[1000:1010]
```

## Palabras comunes irrelevantes

```python
nltk.download("stopwords")
comunes = set(stopwords.words("spanish"))
list(comunes)[-10:]
```

# Palabras sin términos irrelevantes


```python
palabras = []
for palabra in tokens:
  if palabra.isalpha() and palabra not in comunes:
    palabras.append(palabra)
palabras[1000:1010]
```

# Extraer N-gramas más frecuentes
## Ejemplo: Bigramas

```python
ngramas = list(ngrams(palabras, n=2))

# Contar frecuencia de n-gramas
conteos = Counter(ngramas)

# N-gramas más frecuentes
print("\n\nLos 5 n-gramas más frecuentes:\n")
for ngrama, frecuencia in conteos.most_common(n=5):
  print(frecuencia, ngrama)
```
