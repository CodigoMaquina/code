# TF-IDF:
## Frecuencia de Término - Frecuencia Inversa de Documento
## Term Frequency - Inverse Document Frequency

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

documentos = [
  "atleta olímpica gana medalla olímpica",
  "artesano diseña medalla olímpica",
  "comité contrata artesano",
]

# Instancia vectorizador TF-IDF
vectorizador = TfidfVectorizer()

# Vectoriza documentos
matriz_tfidf = vectorizador.fit_transform(documentos)

# Obtiene el nombre de los términos
terminos = vectorizador.get_feature_names_out()

# Imprime vectorización TF-IDF
tabla = pd.DataFrame(matriz_tfidf.todense(),
                     columns=terminos,
                     index=[f"Doc {i+1}" for i in range(len(documentos))])
tabla
```
